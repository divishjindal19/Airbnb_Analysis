```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('dataset.csv.gz')
```

```python
df.head()
```

```python
df.columns
```

```python
# check missing values
print(df.isnull().sum())
```

```python
df.info()
```

```python
df['last review'] = pd.to_datetime(df['last review'],errors= 'coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month' :0, 'last review': df['last review'].min()},inplace=True)
```

```python
df.dropna(subset=['NAME','host name'], inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(columns=["license" , "house_rules"],errors = "ignore")
```

```python
df.head()
```

```python
df['price'] = df['price'].replace(r'[\$,]', '', regex=True).astype(float)
df['service fee'] = df['service fee'].replace(r'[\$,]', '', regex=True).astype(float)
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['price'],bins= 50,kde = True,color='green')
plt.title('Distribution of Listing Price')
plt.xlabel('Price $')
plt.ylabel('Frequency')
plt.show()
```

```python
df['room type']
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x = "room type", data = df, color = 'red')
plt.title("Rooms Distribution")
plt.xlabel("Room Type")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(y = "neighbourhood group",data = df,color = "lightgreen", order = df['neighbourhood group'].value_counts().index)
plt.title("Number of Listing by Neighbourhood Grp")
plt.xlabel("Count")
plt.ylabel("Neighbourhood group")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x = 'room type', y = 'price', hue = 'room type', data = df, palette='Set1')
plt.title("Price V/S Room Type")
plt.xlabel("Room Type")
plt.ylabel("Price $")
plt.legend(title = 'Room type')
plt.show()
```

```python
df['last review'] = pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

plt.Figure(figsize=(12,6))
reviews_over_time.plot(kind = 'line', color = 'blue')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()
```
